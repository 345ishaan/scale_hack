```python
def get_stock_news_in_range(stock_ticker: str, date_after:str, date_before:str, api_key, debug=False):
    query = f'https://api.polygon.io/v2/reference/news?ticker={stock_ticker}&limit=1000&published_utc.gt={date_after}&published_utc.lt={date_before}&sort=published_utc&order=desc&apiKey={api_key}'
    if debug:
        print(query)
    return requests.get(query).json()
    
```

```python
results = get_stock_news_in_range(stock_ticker='AAPL', date_after='2022-01-01', date_before='2022-07-01', api_key='23AKKiIqpATNIGv4O5B_idaf7Ip8WwU4')
```
